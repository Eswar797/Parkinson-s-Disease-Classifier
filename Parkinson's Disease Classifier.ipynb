```python
from fastai.vision import *
from fastai.metrics import *
```

```python
path = '/home/jupyter/Parkinson Disease/Parkinson-s-Disease-Classifier/dataset'
data = ImageDataBunch.from_folder(path, ds_tfms=get_transforms(), valid='test',
                                  size=200, bs=8).normalize(imagenet_stats)
```

```python
data.show_batch(rows=3, figsize=(10,8))
```

```python
print(data.classes)
```

```python
data.label_list
```

```python
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
```

```python
learn.fit_one_cycle(5)
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.recorder.plot_losses()
```

```python
learn.unfreeze()
learn.fit_one_cycle(5, max_lr=slice(1e-04, 1e-04/5))
```

```python
learn.recorder.plot()
```

```python
learn.recorder.plot_losses()
```

```python
interp = ClassificationInterpretation.from_learner(learn)

losses,idxs = interp.top_losses()

len(data.valid_ds)==len(losses)==len(idxs)
```

```python
interp.plot_top_losses(9, figsize=(12,10), heatmap=False)
```
